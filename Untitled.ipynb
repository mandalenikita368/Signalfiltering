```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt

# Simulated noisy signal
t = np.linspace(0, 1.0, 200)
sig = np.sin(2 * np.pi * 5 * t) + 0.5 * np.random.randn(t.size)

# Design a Butterworth low-pass filter
b, a = signal.butter(4, 0.1)  # 4th order, cutoff = 0.1 (normalized)
filtered = signal.filtfilt(b, a, sig)

# Plot
plt.plot(t, sig, label='Noisy Signal')
plt.plot(t, filtered, label='Filtered Signal', linewidth=2)
plt.legend()
plt.title("Signal Filtering with SciPy")
plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```
