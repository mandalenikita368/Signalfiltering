```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import signal

# Generate a noisy sine wave
np.random.seed(0)
t = np.linspace(0, 1, 500, endpoint=False)
signal_clean = np.sin(2 * np.pi * 5 * t)           # 5 Hz sine wave
noise = 0.5 * np.random.normal(size=t.shape)       # Gaussian noise
signal_noisy = signal_clean + noise

# Design a low-pass Butterworth filter
b, a = signal.butter(N=4, Wn=0.1, btype='low')      # 4th order, cutoff = 0.1 (normalized)
signal_filtered = signal.filtfilt(b, a, signal_noisy)

# Plot using seaborn
sns.set(style="darkgrid")
plt.figure(figsize=(12, 6))
sns.lineplot(x=t, y=signal_noisy, label='Noisy Signal', color='red', alpha=0.5)
sns.lineplot(x=t, y=signal_filtered, label='Filtered Signal', color='blue')
sns.lineplot(x=t, y=signal_clean, label='Original Signal', color='green', linestyle='--')
plt.title("Signal Filtering with SciPy and Visualization with Seaborn")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.legend()
plt.tight_layout()
plt.show()
```
